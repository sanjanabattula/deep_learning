```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "loss & accuracy/training_history.xlsx"
train_df = pd.read_excel(file)
train_df
```

```python
# Function to plot loss over epochs
def plot_loss(train_df):
    plt.figure(figsize=(8, 5))
    plt.plot(train_df['Epoch'], train_df['Train Loss'], label='Train Loss', marker='o')
    plt.plot(train_df['Epoch'], train_df['Test Loss'], label='Test Loss', marker='s')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Train vs Test Loss Over Epochs')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot accuracy over epochs
def plot_accuracy(train_df):
    plt.figure(figsize=(8, 5))
    plt.plot(train_df['Epoch'], train_df['Train Accuracy'], label='Train Accuracy', marker='o')
    plt.plot(train_df['Epoch'], train_df['Test Accuracy'], label='Test Accuracy', marker='s')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Train vs Test Accuracy Over Epochs')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the functions to generate the plots
plot_loss(train_df)
plot_accuracy(train_df)
```

```python
# Finetuning data
finetune_file = "loss & accuracy/training_history_finetuning.xlsx"
finetune_df = pd.read_excel(finetune_file)
finetune_df
```

```python
# Function to plot loss (Finetuning data)
def plot_loss(finetune_df):
    plt.figure(figsize=(8, 5))
    plt.plot(finetune_df['Epoch'], finetune_df['Train Loss'], label='Train Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Finetuning Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot accuracy (Finetuning data)
def plot_accuracy(finetune_df):
    plt.figure(figsize=(8, 5))
    plt.plot(finetune_df['Epoch'], finetune_df['Train Accuracy'], label='Train Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Finetuning Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the functions to generate the plots
plot_loss(finetune_df)
plot_accuracy(finetune_df)
```
